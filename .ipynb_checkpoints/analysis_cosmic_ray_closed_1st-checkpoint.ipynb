```python
import numpy as np
from scipy.interpolate import interp1d
from scipy.integrate import simps
import matplotlib.pyplot as plt
import csv
import pandas as pd
from astropy.time import Time
```

```python
data=[]

with open("kailai_second_data.csv", 'r') as file:
  csvreader = csv.reader(file)
  for row in csvreader:
    data.append(row)
```

```python
data_change=[]
index_array=[]
delta_t_array=[]
for i in range(0,len(data)-1):
#for i in range(0,1000):
    index_array.append(i)
    delta_t=24*60*60*(float(data[i+1][1])-float(data[i][1]))
    #print(delta_t)
    delta_t_array.append(delta_t)
```

```python
data_change.append(index_array)
data_change.append(delta_t_array)
```

```python
plt.plot(data_change[0],data_change[1])
plt.xlabel("index")
plt.ylabel("delta_t")
```

```python
plt.plot(data_change[0],data_change[1])
plt.xlabel("index number")
plt.ylabel("delta_t (seconds)")
plt.rcParams["figure.figsize"] = (15,6)
plt.title("Time between Events without the Cover")
```

```python
#Poisson Statistics here
t_avg=0
for i in data_change[1]:
    t_avg+=i

t_avg=t_avg/len(data_change[1])
print("t_avg",t_avg)
```

```python
acc_0=0
acc_1=0
acc_2=0
acc_3=0
acc_4=0
acc_5=0
acc_6=0
acc_7=0
acc_8=0
acc_9=0
acc_10=0
acc_11=0
acc_12=0
for i in data_change[1]:
    if i<1:
        acc_0+=1
    elif i <2:
        acc_1+=1
    elif i<3:
        acc_2+=1
    elif i<4:
        acc_3+=1
    elif i<5:
        acc_4+=1
    elif i<6:
        acc_5+=1
    elif i<7:
        acc_6+=1
    elif i<8:
        acc_7+=1
    elif i<9:
        acc_8+=1
    elif i<10:
        acc_9+=1
    else:
        acc_10+=1

print(acc_0)
print(acc_1)
print(acc_2)
print(acc_5)
print(acc_6)
print(acc_7)
print(acc_8)
print(acc_9)
print(acc_10)
```

```python
total=len(data_change[1])+1
print("total is: ", total)
exps=[0.166695,0.138908,0.115752,0.0964571,0.0803782,0.0669796,0.0558144,0.0465105,0.0387574,0.0322968,0.161451]
for i in range(0,len(exps)):
    exps[i]=total*exps[i]

chi2=0
print("total",total)
exp0=exps[0]
obs0=acc_0
print("expected is: " + str(exp0) + "; observed is: " + str(obs0))
#chi2+=(obs0-exp0)**2/exp0
chi2+=(obs0-exp0)**2/exp0

exp1=exps[1]
obs1=acc_1
print("expected is: " + str(exp1) + "; observed is: " + str(obs1))
chi2+=(obs1-exp1)**2/exp1

exp2=exps[2]
obs2=acc_2
print("expected is: " + str(exp2) + "; observed is: " + str(obs2))
chi2+=(obs2-exp2)**2/exp2

exp3=exps[3]
obs3=acc_3
print("expected is: " + str(exp3) + "; observed is: " + str(obs3))
chi2+=(obs3-exp3)**2/exp3

exp4=exps[4]
obs4=acc_4
print("expected is: " + str(exp4) + "; observed is: " + str(obs4))
chi2+=(obs4-exp4)**2/exp4

exp5=exps[5]
obs5=acc_5
print("expected is: " + str(exp5) + "; observed is: " + str(obs5))
chi2+=(obs5-exp5)**2/exp5

exp6=exps[6]
obs6=acc_6
print("expected is: " + str(exp6) + "; observed is: " + str(obs6))
chi2+=(obs6-exp6)**2/exp6

exp7=exps[7]
obs7=acc_7
print("expected is: " + str(exp7) + "; observed is: " + str(obs7))
chi2+=(obs7-exp7)**2/exp7

exp8=exps[8]
obs8=acc_8
print("expected is: " + str(exp8) + "; observed is: " + str(obs8))
chi2+=(obs8-exp8)**2/exp8

exp9=exps[9]
obs9=acc_9
print("expected is: " + str(exp9) + "; observed is: " + str(obs9))
chi2+=(obs9-exp9)**2/exp9

exp10=exps[10]
obs10=acc_10
print("expected is: " + str(exp10) + "; observed is: " + str(obs10))
chi2+=(obs10-exp10)**2/exp10

print("chi2_total: ", chi2)
```
